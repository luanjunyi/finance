```python
from valuesider_crawler.portfolio_crawler import PortfolioCrawler
```

```python
crawler = PortfolioCrawler('https://valuesider.com/guru/howard-marks-oaktree-capital-management/portfolio/2013/2')
d = await crawler.crawl()

for k, v in d.items():
    v['period'] = k

df = pd.concat(d.values()).sort_values(by='period')
df.to_csv('oaktree-portfolio-history.csv', index=False)
```

```python
trmd = df[df.symbol == 'TRMD'].copy()
trmd['share_diff'] = trmd.shares.pct_change() * 100
trmd['share_in_m'] = trmd.shares / 1000000
```

```python
import logging
logging.basicConfig(level=logging.INFO)
plt.figure(figsize=(12, 5))
plt.plot(trmd.period, trmd.share_in_m, 'x-')
plt.xticks(rotation=75)
plt.ylabel('Shares (Millions)')
_ = plt.title('OAKTREE TRMD Shares in Millions')
```
