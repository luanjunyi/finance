```python
import pandas as pd
from fmp_data import Dataset, AFTER_PRICE

ap = ["GOOGL", "CLS", "SYF", "ZETA", "EAT", "SKYW", "STRL", "GM", "TMUS", "TWLO", "PPC", "CRM", "RCL", "MFC", "ATGE", "MHO", "MHO", "ANF", "BLBD", "URI", "CAAP", "AMPH", "APP", "AGX", "SFM", "POWL", "CAH", "UBER", "CMCSA", "BRK.B", "OKTA", "PEP", "MOD", "MOD", "RGA", "GRBK", "TEX", "QTWO"]
d = Dataset(ap, {'weighted_average_shares_outstanding_diluted': 'num_shares', AFTER_PRICE: 'price'}).data
d['date'] = pd.to_datetime(d['date'])
d = d.sort_values(by='date', ascending=False)
d = d.drop_duplicates(subset='symbol', keep='first')
d['market_cap_in_billions'] = d['price'] * d['num_shares'] / 1000000000
d = d.sort_values(by='market_cap_in_billions', ascending=False)
```
